```python
# Importing libraries
from tkinter import *
from tkinter import filedialog
from bing_image_downloader import downloader

# Creating a window
window = Tk()
window.title("Bing Image Downloader")
window.geometry("400x200")

# Creating labels and entries
keyword_label = Label(window, text="Keyword:")
keyword_entry = Entry(window)
amount_label = Label(window, text="Amount:")
amount_entry = Entry(window)

# Creating a function to ask for save location
def ask_location():
    # Opening a dialog box for the user to choose a folder
    location = filedialog.askdirectory()
    # Displaying the chosen folder in the entry
    location_entry.delete(0, END)
    location_entry.insert(0, location)

# Creating a button and an entry for save location
location_button = Button(window, text="Browse", command=ask_location)
location_entry = Entry(window)

# Creating a function to download images
def download_images():
    # Getting the parameters from the entries
    keyword = keyword_entry.get()
    amount = int(amount_entry.get())
    location = location_entry.get()
    
    # Downloading images using bing-image-downloader
    downloader.download(keyword, limit=amount,
                        output_dir=location,
                        adult_filter_off=True,
                        force_replace=False)
    
    # Showing a message when done
    message_label.config(text="Download complete!")

# Creating a button and a message label
download_button = Button(window, text="Download", command=download_images)
message_label = Label(window)

# Placing the widgets on the window using grid layout
keyword_label.grid(row=0,column=0,padx=10,pady=10)
keyword_entry.grid(row=0,column=1,padx=10,pady=10)
amount_label.grid(row=1,column=0,padx=10,pady=10)
amount_entry.grid(row=1,column=1,padx=10,pady=10)
location_button.grid(row=(2),column=(0),padx=(5),pady=(5))
location_entry.grid(row=(2),column=(1),padx=(5),pady=(5))
download_button.grid(row=(3),columnspan=(2),padx=(5),pady=((20),(5)))
message_label.grid(row=(4),columnspan=(2),padx=(5),pady=((5)))

# Running the main loop of the window
window.mainloop()
```
