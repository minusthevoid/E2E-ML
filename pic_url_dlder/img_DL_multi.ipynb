```python
# Import libraries
import os
import requests
from bing_image_downloader import downloader
import PySimpleGUI as sg
# Import glob module
import glob

# Define layout for GUI window
layout = [
    [sg.Text('Enter one or more keywords to search (separated by commas):')],
    [sg.Input(key='search_terms')],
    [sg.Text('Enter number of images to download per keyword:')],
    [sg.Input(key='num_images')],
    [sg.Text('Select folder name to save images:')],
    # Use FolderBrowse() function with key parameter
    [sg.Input(key='output_dir'), sg.FolderBrowse()],
    [sg.Button('Download'), sg.Button('Exit')]
]

# Create GUI window
window = sg.Window('Image Downloader', layout)

# Loop through events and values from window
while True:
    event, values = window.read()
    
    # If user clicks Exit or closes window, break loop
    if event == 'Exit' or event == sg.WIN_CLOSED:
        break
    
    # If user clicks Download, get inputs and download images
    if event == 'Download':
        # Get inputs from values dictionary
        search_terms = values['search_terms'].split(',')
        num_images = int(values['num_images'])
        output_dir = values['output_dir']
        
        # Loop through each keyword and download images 
        for term in search_terms:
            # Remove leading and trailing spaces from keyword 
            term = term.strip()
            # Create a subdirectory for each keyword using os.mkdir with output directory and keyword as arguments 
            sub_dir = os.path.join(output_dir, term)
            os.mkdir(sub_dir)
            # Download images using bing_image_downloader library with subdirectory as output directory 
            downloader.download(term, limit=num_images, output_dir=sub_dir)
            # Get the list of downloaded image paths using glob.glob with pattern matching keyword and subdirectory 
            image_paths = glob.glob(os.path.join(sub_dir, term + '*', '*'))
            # Rename each image file with keyword prefix 
            for i, path in enumerate(image_paths):
                new_path = os.path.join(sub_dir, term + '_' + str(i) + os.path.splitext(path)[1])
                os.rename(path, new_path)
        
        # Show a popup message when done downloading images 
        sg.popup(f'Downloaded {len(search_terms) * num_images} images successfully!')

# Close window when loop ends        
window.close()
```
